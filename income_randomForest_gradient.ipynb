```python
label = preprocessing.LabelEncoder() 
  
income[' workclass']= label.fit_transform(income[' workclass']) 
income[' education']= label.fit_transform(income[' education']) 
income[' relationship']= label.fit_transform(income[' relationship']) 
income[' marital-status']= label.fit_transform(income[' marital-status']) 
income[' occupation']= label.fit_transform(income[' occupation']) 
income[' race']= label.fit_transform(income[' race']) 
income[' native-country']= label.fit_transform(income[' native-country']) 
income[' income']= label.fit_transform(income[' income'])  
```

```python
income.head()
```

```python
one_hot = pd.get_dummies(income[' sex'])
income = income.drop(' sex',axis = 1)
income = income.join(one_hot)
```

```python
income.head()
```

```python
income_majority = income[income[" income"]==0]
income_minority = income[income[" income"]==1]
 

income_minority_upsampled = resample(income_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=20000,    # to match majority class
                                 random_state=100) # reproducible results
 

income = pd.concat([income_majority, income_minority_upsampled])
income[" income"].value_counts()
```

```python
y=income[" income"].values.reshape(-1,1)
x=income.drop([" income" , " education-num"], axis=1).values
```

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
x=pd.DataFrame(data=x)
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print("x train: ",x_train.shape)
print("x test: ",x_test.shape)
print("y train: ",y_train.shape)
print("y test: ",y_test.shape)
```

```python
def model(model):
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    print("train score",model.score(x_train,y_train) )
    print("test score" , model.score(x_test,y_test))
    print(classification_report(y_test,pred))
```

```python
random = RandomForestClassifier(max_depth=27, random_state=0 , criterion='entropy')
```

```python
model(random)
```

```python
grad=GradientBoostingClassifier(max_depth=10)
```
